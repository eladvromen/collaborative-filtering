```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("edited_rating.csv")
```

```python
df.shape
```

```python
num_users = df["userId"].max()
num_movies = df["movie_idx"].max()
print("users", num_users, "movies", num_movies)
```

```python
df.head(10)
```

```python
num_of_users = 7500
num_of_movies = 1500
user_filter = df['userId'].value_counts()[:num_of_users].index.tolist()
movie_filter = df['movie_idx'].value_counts()[:num_of_movies].index.tolist()
```

```python
df_small= df[df["userId"].isin(user_filter) & df["movie_idx"].isin(movie_filter)].copy()
```

```python
df_small.head()
```

```python
user_ids = set(df_small.userId.values)
movie_ids= set(df_small.movie_idx.values)
new_user_id_map = {}
new_movie_id_map = {}
i=0
j=0
for old in user_ids:
  new_user_id_map[old] = i
  i+=1

for old in movie_ids:
  new_movie_id_map[old] = j
  j += 1
```

```python
new_user_id_map 
```

```python
df_small.head()
```

```python
df_small['new_userId'] = df["userId"].map(new_user_id_map)
df_small['new_movieId'] = df["movie_idx"].map(new_movie_id_map)
```

```python
df_small.tail(10)
```

```python
print("max user id:", df_small.new_userId.max())
print("max movie id:", df_small.new_movieId.max())

print("small dataframe size:", len(df_small))
```

```python
df_small.to_csv('small_rating.csv', index=False)
```
