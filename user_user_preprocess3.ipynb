```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import pickle
```

```python
df = pd.read_csv("small_rating.csv")
```

```python
df_train, df_test = train_test_split(df, test_size=0.2, random_state=42)
```

```python
train = df_train[["new_movieId", "new_userId", "rating"]]
test = df_test[["new_movieId", "new_userId", "rating"]]
user2movies = {}#mapping all movies rated by user to him
movie2users = {}#mapping all useres who rated the movie to him
usermovie2rating = {}#mapping he rating to each user-movie coupling
usermovie2rating_test = {}
count = 0
```

```python
def update_dicts(row):
  global count
  count += 1
  if count % 100000 == 0:
    print("processed:")

  i = int(row.new_userId)
  j = int(row.new_movieId)
  if i not in user2movies:
    user2movies[i] = [j]
  else:
    user2movies[i].append(j)

  if j not in movie2users:
    movie2users[j] = [i]
  else:
    movie2users[j].append(i)

  usermovie2rating[(i,j)] = row.rating
```

```python
train.apply(update_dicts, axis=1)
```

```python
def update_dict_test(row):
    global count
    count += 1
    if count % 100000 == 0:
       print("processed:")

    i = int(row.new_userId)
    j = int(row.new_movieId)
    
    usermovie2rating_test [(i,j)] = row.rating
```

```python
count = 0
test.apply(update_dict_test, axis=1)
```

```python
with open('user2movie.json', 'wb') as f:
  pickle.dump(user2movies, f)

with open('movie2user.json', 'wb') as f:
  pickle.dump(movie2users, f)

with open('usermovie2rating.json', 'wb') as f:
  pickle.dump(usermovie2rating, f)

with open('usermovie2rating_test.json', 'wb') as f:
  pickle.dump(usermovie2rating_test, f)
```
